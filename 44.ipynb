```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

X,y = make_regression(n_samples=300, n_features=1)
x1 = X
y1 = y

print(X.shape)

#1. 초기값 구하기
```

```python
W = np.random.rand(2)

one = np.ones([300,1])

X= np.append(X, one, axis=1)

hypothesis=np.dot(X,W)
```

```python
#2. 초기값의 cost 구하기

cost=np.mean(np.square(hypothesis-y))
print(cost)

#3. W값 update

Gradient=np.dot(2*(np.dot(X,W)-y),X)
print(Gradient)

```

```python
def cost(W):
    return np.mean(np.square(np.dot(X,W)-y))

for i in range(1000):
    W = W - 0.001*Gradient
    Gradient=np.dot(2*(np.dot(X,W)-y),X) # np.dot(X,W)

print(W, cost(W))
```
